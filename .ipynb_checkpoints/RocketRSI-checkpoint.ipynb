```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, date
```

```python
daily_ry = yf.download("RY.TO", start = datetime(2024,1,1), end = date.today(), interval = '1d')
```

```python
def rsi(prices, period=10):
    
    prices = np.array(prices)
    rsi = np.zeros_like(prices)
    
    deltas = np.diff(prices)
    prices_up = deltas*(np.sign(deltas) + 1)/2
    prices_down = -deltas*(1 - np.sign(deltas))/2
    
    
    for i in range(len(prices)):
        
        if i < period - 1:
            
            rsi[i] = np.nan
            
        else:
            
            rs = np.sum(prices_up[i - period+1:i+1])
            rs /= np.sum(prices_down[i - period+1:i+1]) + 1e-10
            rsi[i] = 100 - 100/(1+ rs)
            
    return rsi
```

```python
def myrsi(prices, period=10):
    
    prices = np.array(prices)
    myrsi = np.zeros_like(prices)
    
    deltas = np.diff(prices)
    prices_up = deltas*(np.sign(deltas) + 1)/2
    prices_down = -deltas*(1 - np.sign(deltas))/2
    
    
    for i in range(len(prices)):
        
        if i < period - 1:
            
            myrsi[i] = np.nan
            
        else:
            
            CU = np.sum(prices_up[i - period+1:i+1])
            CD = np.sum(prices_down[i - period+1:i+1]) 
            myrsi[i] = (CU - CD)/(CU + CD)
            
    return myrsi    
```

```python
def SuperSmootherFilter(prices, smoothlength = 8):
    
    Filt = np.zeros_like(prices)
    
    a1 = np.exp(-np.sqrt(2)*np.pi/smoothlength)
    b1 = 2*a1*np.cos(np.sqrt(2)*np.pi/smoothlength)
    c2 = b1
    c3 = -a1*a1
    c1 = 1 - c2 - c3
    
    for i in range(len(prices)):
        
        if i < 3:
            Filt[i] = prices[i]
            
        else:
            
            Filt[i] = c1*(prices[i] + prices[i-1])/2
            Filt[i] += c2*Filt[i-1] + c3*Filt[i-2]
    return Filt
```

```python
def momentum(prices, period = 10):
    
    prices = np.array(prices)
    momentum = np.zeros_like(prices)
    
    for i in range(len(prices)):
        
        if i < period - 1:
            
            momentum[i] = 0.
            
        else:
            
            momentum[i] = prices[i] - prices[i - period]
            
    return momentum
```

```python
## Graphique de l'indicateur RSI sur le cours de l'indice RY

fig, axs = plt.subplots(2,1,figsize=(16,9), gridspec_kw={'height_ratios': [2, 1]})

fig.suptitle("Indicateur RSI sur le cours de l'indice RY", fontsize = 20)
axs[0].plot(daily_ry.index, daily_ry['Close'])


axs[1].set_ylim(0,100)
axs[1].axhline(y=30, color = 'r', ls = '--')
axs[1].axhline(y=70, color = 'r', ls = '--')
axs[1].plot(daily_ry.index, rsi(daily_ry['Close']))


plt.show()
```

```python
## Graphique de l'indicateur MyRSI sur le cours de l'indice RY

smooth_prices = SuperSmootherFilter(daily_ry['Close'])
myrsi_prices = myrsi(smooth_prices)

fig, axs = plt.subplots(2,1,figsize=(16,9), gridspec_kw={'height_ratios': [2, 1]})

fig.suptitle("Indicateur MyRSI sur le cours de l'indice RY", fontsize = 20)
axs[0].plot(daily_ry.index, daily_ry['Close'])


axs[1].set_ylim(-2,2)

axs[1].plot(daily_ry.index, myrsi_prices)


plt.show()
```

```python
## Graphique de l'indicateur MyRSI sur le cours de l'indice RY

prices = np.array(daily_ry['Close'])
mom_prices = momentum(prices) 
smooth_prices = SuperSmootherFilter(mom_prices)
myrsi_prices = myrsi(smooth_prices)

for i in range(len(myrsi_prices)):
    
    if myrsi_prices[i] > 0.999:
        myrsi_prices[i] = 0.999
        
    if myrsi_prices[i] < -0.999:
        myrsi_prices[i] = -0.999
rocketrsi = 1/2*np.log((1+myrsi_prices)/(1-myrsi_prices))

fig, axs = plt.subplots(2,1,figsize=(16,9), gridspec_kw={'height_ratios': [2, 1]})

fig.suptitle("Indicateur RocketRSI sur le cours de l'indice RY", fontsize = 20)
axs[0].plot(daily_ry.index, daily_ry['Close'])
axs[1].axhline(y=-2, color = 'r', ls = '--')
axs[1].axhline(y=2, color = 'r', ls = '--')

axs[1].plot(daily_ry.index, rocketrsi)


plt.show()
```
